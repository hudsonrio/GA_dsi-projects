```python
import pandas as pd 

fent = pd.read_csv('/Users/HudsonCavanagh/Dropbox/Capstone/fent_data/fent_mod.csv')
fent = fent.T
# fent.head()
fent.columns = fent.T.iloc[:,0]
# fent = fent.drop(['State'])
fent.iloc[:,5:]


def range_fixer(item):
    item_avg = []
    item = str(item)
    if "to" in item:
        a = item.split(" to ")[0]
        b = item.split(" to ")[1]
        item_avg.append(float(a))
        item_avg.append(float(b))
        a_1 = sum(item_avg)
        b_1 = len(item_avg)
        print (a_1/b_1)
        return (a_1/b_1)
    elif "no reports" in item:
        return 0

fent.loc['2013'] = fent.loc['2013'].apply(lambda x: range_fixer(x))                    
fent.loc['2014'] = fent.loc['2014'].apply(lambda x: range_fixer(x))
fent.loc['2013', 'USA(fent)'] = 934 # this should have been on source doc: https://docs.google.com/spreadsheets/d/1RWCVU7gapVDp3m54aJ30jIYkpoQ6AnoLHeb0gE_2bjI/edit#gid=1118416318
fent.loc['2013', 'USA(op)'] = 270981
#only have half of 2014 data available, so assuming that 0.49437% of fent and ops are discovered in second half of year, based on prior years
fent.loc['2014', 'USA(op)'] = 279704 #see above for methodology, from http://www.deadiversion.usdoj.gov/nflis/spec_rpt_opioids_2014.pdf
fent.loc['2014', 'USA(fent)'] = 2032 #includes both fentanyl and acetyl fentanyl, which was first reported in this year

```
