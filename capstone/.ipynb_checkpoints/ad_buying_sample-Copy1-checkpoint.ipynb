```python
import pandas as pd
import numpy as np
import sklearn


df = pd.read_csv('/Users/HudsonCavanagh/Documents/Q1_viewership_data.csv', encoding='utf-8')
df = df.T
df = df.applymap(lambda x: x.replace("\t", ""))
```

```python
df.columns = df.loc["Unnamed: 0",:]
df.drop("Unnamed: 0", inplace=1)

df #looks clean for now
```

### Belmont Cadillacâ€™s Memorial Day Sale -- $10,000 budget
because we know that two people cannot watch simultaneous channels, we want to spread out programming if and where possible

however, time does not otherwise matter --> going to change this to 4 columns instead



```python
a = df.iloc[:,0:4]
b = df.iloc[:,4:8]
c = df.iloc[:,8:12]
d = df.iloc[:,12:16]
```

```python
a.head()
col_list = ["show", 'target_viewers', 'available_spots', 'price_per_spot']
a.columns = col_list
```

```python
b.head()
b.columns = col_list
```

```python
c.head()
c.columns = col_list
```

```python
d.head()
d.columns = col_list
```

```python
views = pd.concat([a,b,c,d], axis=0) #this works 1) need to make floats 2) need more data
views.head()
```

```python
def floater(example):
    example = example.replace(",", "")
    try:
        example = str(example).split('$')[1]
        return float(example)
    except:
        return float(example)
    
views['target_viewers'] = views['target_viewers'].apply(lambda x: floater(x))
views['price_per_spot'] = views['price_per_spot'].apply(lambda x: floater(x))
views['available_spots'] = views['available_spots'].apply(lambda x: float(x))


views['target_viewers']     
    
```

```python
views['viewers_per_dollar'] = views['target_viewers']/ views['price_per_spot']
```

```python
import seaborn as sns
%matplotlib inline
sns.distplot(views['viewers_per_dollar'])
```

### We want to make sure we're targeting as many opportunities on the right of the above graph as possible: need to maximize viewers per dollar

```python
views = views.sort_values(by='viewers_per_dollar', ascending=0)
```

```python
views['max_spending'] = views['available_spots']*views['price_per_spot']
views['advisable_available_spots'] = views['available_spots'].apply(lambda x: 3 if x>5 else x)
#by playing with the line above, we can change our assumption about maximum "advisable" spots

views['advised_spending'] = views['advisable_available_spots'] *views['price_per_spot']
```
