```python
import pandas as pd
import numpy as np
#downloaded from https://www.census.gov/2010census/csv/pop_density.csv -- state level

pop_dense = pd.read_csv('/Users/HudsonCavanagh/Dropbox/Capstone/non_cdc_data/pop_density.csv')
pop_dense.head()
```

```python
pop_dense_cols = pop_dense.iloc[2,:]
pop_dense = pop_dense.iloc[3:,:]
pop_dense.columns = pop_dense_cols 
# pop_dense.head()
pop_dense.columns
```

```python
def floater(item):
    item = str(item)
    item = item.replace(",", "")
    try:
        item = float(item)
    except:
        pass 
    return item

pop_dense.iloc[:,1:] = pop_dense.iloc[:,1:].applymap(lambda x: floater(x))

pop_dense['2005_dense'] = (pop_dense['2000_DENSITY'] + pop_dense['2010_DENSITY'])/2.0
```
