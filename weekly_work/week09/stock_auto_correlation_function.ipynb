```python
from collections import defaultdict
auto_corr_dict = defaultdict()
lag_list = [7,14,30,91,100,182,364, 500, 700, 900, 1100,1300, 1500]

def auto_corr(stock, lag_list, close=1):
    auto_corr_list = []
    for h in lag_list:
        auto_corr_list.append(stock['Close'].autocorr(lag=h))
    auto_corr = pd.DataFrame(auto_corr_list)
    return(auto_corr)
   

att_corr = auto_corr(att, lag_list)
vb_corr =auto_corr(vb, lag_list)
tmob_corr =auto_corr(tmob, lag_list)
sprint_corr =auto_corr(sprint, lag_list)
usm_corr =auto_corr(usm, lag_list)




cors = pd.concat([att_corr,vb_corr,tmob_corr, sprint_corr, usm_corr], axis=1)
# cors = cors.index([7,14,30,182,364])
cors.columns = ['att', 'verizon', 'tmobile', 'sprint', 'usm']
cors = cors.T
cors.columns = lag_list
cors = cors.T
cors.head(15)

# att['Close'].autocorr(lag=7)
```
