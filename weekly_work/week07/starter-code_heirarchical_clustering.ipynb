```python
from sklearn import preprocessing


def make_nums(x):
    if x=='Iris-setosa':
        return(-1)
    elif x=='Iris-virginica':
        return(0)
    elif x=='Iris-versicolor':
        return(1)
    
iris['type_num'] = iris['Name'].apply(make_nums)


```

```python
iris.drop('Name', axis=1, inplace=1)
iris_scale = preprocessing.StandardScaler().fit_transform(iris)
iris_scale = pd.DataFrame(iris_scale, columns=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'scaled_type'])
```

## Part 2: Hierarchical Clustering

Now, let's perform the actual clustering on our set

```python
from __future__ import print_function, division

from IPython.core.display import Image, HTML

import numpy as np
import pandas as pd

from scipy.cluster import hierarchy

import matplotlib.pyplot as plt
%matplotlib inline


Z = hierarchy.linkage(X, 'complete')
fig, ax = plt.subplots(figsize=(16,12))
dn = hierarchy.dendrogram(Z, labels=iris_scale.index)

max_dist = 5
clusters = hierarchy.fcluster(Z, max_dist, criterion='distance')


# fig, ax = plt.subplots(figsize=(12,8))
# dn = hierarchy.dendrogram(Z, orientation='left', labels=X.index)
```

```python
fig, ax = plt.subplots(figsize=(16,12))
plt.scatter(iris_scale.iloc[:,0], iris_scale.iloc[:,0], c=clusters, cmap='prism', s=100);
```

```python
Z2 = hierarchy.linkage(X, 'single')
fig1, ax1 = plt.subplots(figsize=(16,12))
dn2 = hierarchy.dendrogram(Z2, labels=iris_scale.index)

max_dist_2 = 10
clusters = hierarchy.fcluster(Z2, max_dist_2, criterion='distance')
```

```python
Z3 = hierarchy.linkage(X, 'average')
fig1, ax1 = plt.subplots(figsize=(16,12))
dn3 = hierarchy.dendrogram(Z3, labels=iris_scale.index, truncate_mode='lastp', p=15)

max_dist_3 = 15
clusters = hierarchy.fcluster(Z3, max_dist_3, criterion='distance')
```
