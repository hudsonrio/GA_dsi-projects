## Step 3: Define the Variables

Next, let's define the x and y variables: Airport is going to be our target variable

```python
# airport_dums = pd.get_dummies(airport['airport'])

y = airports['airport']
X = airports.iloc[:,1:]
# X.drop('airport', axis=1, inplace=1)
```

```python
airports.head()
```

## Step 4: Covariance Matrix

Then, standardize the features for analysis

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import math
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
from sklearn.cluster import KMeans
from sklearn import metrics


X_scale = StandardScaler().fit_transform(X)
```

Next, create the covariance matrix from the standardized features and decompose these values to find the eigenvalues and eigenvectors

```python


cov_matrix = np.cov(X_scale.T)

```

# Step 5: Eigenpairs

Then, check your eigenvalues and eigenvectors:

```python
eigenValues, eigenVectors = np.linalg.eig(cov_matrix)
```

To find the principal componants, find the eigenpairs, and sort them from highest to lowest. 

```python
# pair = []
# for h in range(len(eigenVectors)):
#     eigenPair = np.abs(eigenValues[h], eigenVectors[:,h])
#     pair.append(eigenPair)
    
# eigenPairs = pair
# print(eigenPairs)

eigenPairs = [(np.abs(eigenValues[i]), eigenVectors[:,i]) for i in range(len(eigenValues))]

    
eigenPairs.sort()
eigenPairs.reverse()
for i in eigenPairs:
    print(i[0])
```

## Step 6: Explained Variance

```python
totalEigen = sum(eigenValues)
varExpl = [(i / totalEigen)*100 for i in sorted(eigenValues, reverse=True)]
```

```python
varExpl
```

Now, calculate the explained variance and the cumulative explained variance

```python
cum_variance = np.cumsum(varExpl)
cum_variance


range(len(cum_variance))
```

```python
sns.barplot(range(len(cum_variance)), cum_variance, palette="BuGn_d")
sns.set(style="white", context="talk")
plt.ylim(0,100)
```

** What does the explained variance tell us?**: Here, we can see that 81.77% of the behavior can be explained by the first two principal components

## Step 7: Perform the PCA

Instead of creating the projection matrix, we're going to use Scikit's built in function. Now that we have discovered the principal componants, we have an educated idea on how many componants to pass to the function. 

```python
pca_2 = PCA(n_components=2)
air_pca_2_features = pca_2.fit_transform(X_scale) # Use the eigenvectors to transform xStand onto new axes
```

Create a dataframe from the PCA results

```python
air_pca_2_features = pd.DataFrame(air_pca_2_features)
air_pca_2_features.head()
```

Now, create a new dataframe that uses the airport and year from the original set and join the PCA results with it to form a new set

```python
df = pd.merge(air_pca_2_features, airports,left_index=1, right_index=1)

len(df)
```

```python
df
df_basic = df.iloc[:,:4]
df_basic.columns = ['pca1', 'pca2', 'airport', 'year']
df_basic
```

Next, graph the results onto the new feature space

```python
from pylab import rcParams

rcParams['figure.figsize'] = (12, 20)
graph = sns.regplot('pca1','pca2',data=df_basic, fit_reg=1)


for i, airport in enumerate(df_basic['airport']):
    graph.annotate(airport, (df_basic.iloc[i].pca2, df_basic.iloc[i].pca1))
```
