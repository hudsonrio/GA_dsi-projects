```python
airports = pd.read_csv('/Users/HudsonCavanagh/Documents/airport2.csv')
air = pd.DataFrame(airports)
air.head()
```

Convert the class labels, in this case the airports, to numeric. *Remember*, we've truncated the airport dataset for simplicty to make this part easy! 

```python
def air_to_num(string):
    if string == "ANC":
        return(0)
    if string == "ATL":
        return(1)
    if string == "ABQ":
        return(2)
    
air['airport_num'] = air['Airport'].apply(air_to_num)
```

## 3.  Perform a K-Means Clustering 

#### 3.1 Split the dataset into the attribute data and class labels

```python
X = air.loc[:,["Departure Cancellations", "Arrival Cancellations", "Departure Diversions", "Arrival Diversions"]]
y = air.loc[:,"airport_num"]
```

```python
any([np.any(np.isnan(X)), np.any(np.isnan(y))])
```

```python
from sklearn import preprocessing

X_scale = preprocessing.StandardScaler().fit_transform(X)
X_scale = pd.DataFrame(X_scale)
```

#### 3.2 Conduct the k-means clustering

```python
k = 3
kmeans = cluster.KMeans(n_clusters=k)
kmeans.fit(X_scale)
```

#### 3.3 Compute the labels and centroids

```python
labels = kmeans.labels_
labels
```

```python
centroids = kmeans.cluster_centers_
centroids
```

```python
s = []
for k in range(2, 20):
    kmeans = cluster.KMeans(n_clusters=k)
    kmeans.fit(X_scale)
    labels = kmeans.labels_
    s_i = metrics.silhouette_score(X_scale, labels, metric='euclidean')
    s.append(s_i)
plt.plot(range(2,20), s)
plt.ylabel("Silouette")
plt.xlabel("k")
```

#### 3.4 Format the resulting predicted "y" variables and check the accuracy score, classification report, and confusion matrix

```python
from sklearn import metrics
print(metrics.score(y, labels, scoring='f1'))
```

```python
print(metrics.classification_report(y, labels))
```

```python
print(metrics.confusion_matrix(y, labels))
```

## 6. Hierarchical Clustering

#### 6.1 Create the linkage for clustering

```python
X_mat = X_scale.as_matrix(columns=None)
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
%matplotlib inline


Z = hierarchy.linkage(X_mat, 'complete')
fig, ax = plt.subplots(figsize=(16,12))
dn = hierarchy.dendrogram(Z, labels=X_scale.index)

max_dist = 5
clusters = hierarchy.fcluster(Z, max_dist, criterion='distance')
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
%matplotlib inline


Z1 = hierarchy.linkage(X_mat, 'single')
fig1, ax1 = plt.subplots(figsize=(16,12))
dn1 = hierarchy.dendrogram(Z1, labels=X_scale.index)

max_dist_1 = 5
clusters = hierarchy.fcluster(Z1, max_dist_1, criterion='distance')
```

```python
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt
%matplotlib inline


Z2 = hierarchy.linkage(X_mat, 'average')
fig2, ax2 = plt.subplots(figsize=(16,12))
dn2 = hierarchy.dendrogram(Z1, labels=X_scale.index)

max_dist_2 = 4
clusters = hierarchy.fcluster(Z2, max_dist_2, criterion='distance')
```
