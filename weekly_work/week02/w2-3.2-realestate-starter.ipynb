```python
# The simplest pivot table must have a dataframe and an 
# index . Let’s use the city as our index.

pivot1 = pd.pivot_table(realestate, values=['baths', 'beds', 'price', 'sq__ft'],index='city', margins=1)

pivot1
```

```python
# How about indexing on multiple values. Let's look at the 
# data by city and zip next. 

pivot2 = pd.pivot_table(realestate, values=['baths', 'beds', 'price', 'sq__ft'],index=['city','zip'], margins=1)
pivot2
```

```python
# What we really want to take a look at next is price. So, 
# let's explicitly define the 'price' column by 
# using the values field.

pivot3 = pd.pivot_table(realestate, values=['price', 'sq__ft'],index=['city','zip'], margins=1)
pivot3
```

```python
# The 'price' column automatically averages the data.  
# We can do a count or a sum by using aggfunc and np.sum.

pivot4 = pd.pivot_table(realestate, values=['baths', 'beds', 'price', 'sq__ft'],index=['city','zip'], margins=1, aggfunc=np.sum)
pivot4
```
