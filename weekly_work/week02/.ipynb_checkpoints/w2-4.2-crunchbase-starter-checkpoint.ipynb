#### 1. Clean the data. Use .apply and lambda functions where you can

```python
# rename the columns to remove spaces and extra characters

df = df_early
#get rid of spaces in columns
df.rename(columns=lambda x: x.strip(), inplace=True)
#split the category_list into a real list
df['category_list'] = df['category_list'].apply(lambda x: str(x).split('|')[1:-1])
#get rid of all whitespace before and after all elements of dataset
df.applymap(lambda x: str(x).strip())
#string.translate(s, table[, deletechars])
#['#','&','.','[',']','-','+','?','!','(',')','{','}','/','$']))
##&.,[]-+?!(){}/$
#can't index by name beacuse its messy: 
df.index = df["name"]
#df_cleaner = df
df.head()
#df_cleaner.head()
```

```python
# convert the funding, year founded, and funding_rounds columns to integers
#find the weird blank value space at index 9, then replace it with np.nans


def _dollars_to_int(x):
    try:
        y = (str(x)).replace(',', '')
        return int(y)
    except:
        pass
    
def date_to_dtime(x):
    try:
        int(x).to_datetime
    except:
        pass


df['founded_at'].apply(date_to_dtime)

custom_null = df['funding_total_usd'][9]


#funding amount conversion
#df['funding_total_usd'] = df['funding_total_usd'].map(lambda x: str(x).strip(","))
df['funding_total_usd'] = df['funding_total_usd'].map(_dollars_to_int)
df['funding_rounds'] = df['funding_rounds'].map(_dollars_to_int)
#year founded conversation

#funding_rounds
df['funding_rounds'].map(lambda x: int(x))
#funding per round creation

#print(df['funding_total_usd'])


df['funding_total_usd'].replace(custom_null, np.NaN, inplace=True)
df['avg_funding_per_round'] = df['funding_total_usd']/df['funding_rounds']
#df['avg_funding_per_round'].head()



    

    
```
