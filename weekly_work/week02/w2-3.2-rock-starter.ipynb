```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Read in rock.csv as a data frame
rock = pd.read_csv('/Users/HudsonCavanagh/dsi-projects/DSI-NYC-1/curriculum/week-02/3.2-lab/assets/datasets/rock.csv')
type(rock)
```

```python
rock.head()
```

```python
# The simplest pivot table must have a dataframe and an 
# index . Let’s use the ARTIST CLEAN as our index.

rock_pivot = pd.pivot_table(rock, index='ARTIST CLEAN', margins=1)
rock_pivot
```

```python
# How about indexing on multiple values. Let's look at the 
# data by ARTIST CLEAN and Release Year next. 

rock_pivot2 = pd.pivot_table(rock, index=['ARTIST CLEAN','Release Year'], margins=1)
rock_pivot2
```

```python
# What we really want to take a look at next is PlayCount. So, 
# the F*G, First?, and Year? columns aren’t useful. Let’s 
# remove them and explicitly define the 'PlayCount' column by 
# using the values field.

rock_pivot3 = pd.pivot_table(rock, index=['ARTIST CLEAN','Release Year'], values="PlayCount", margins=1)
rock_pivot3
```

```python
# The 'PlayCount' column automatically averages the data.  
# We can do a count or a sum by using aggfunc and np.sum.

rock_pivot4 = pd.pivot_table(rock, index=['ARTIST CLEAN','Release Year'], values="PlayCount", margins=1, aggfunc=(np.sum))
rock_pivot4
```
