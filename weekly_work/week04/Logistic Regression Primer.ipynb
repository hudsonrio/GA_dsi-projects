```python
import statsmodels.api as sm
import pandas as pd
import pylab as pl
import numpy as np

df = pd.read_csv("http://www.ats.ucla.edu/stat/data/binary.csv")
df.columns = list(df.columns)[:3] + ["prestige"]

```

```python
dummy_ranks = pd.get_dummies(df["prestige"], prefix="prestige") #really powerful

print(dummy_ranks)
```

```python
cols_to_keep = ["admit", "gre", "gpa"]
data = df[cols_to_keep].join(dummy_ranks.ix[:, "prestige_2":])


```

```python
data
```

```python
data["intercept"] = 1.0
#THERE MUST ALWAYS BE A COLUMN WITH 1s with STATS MODELS
train_cols = data.columns[1:]

```

```python
logit = sm.Logit(data["admit"], data[train_cols])

result = logit.fit()
result.summary2()
```
