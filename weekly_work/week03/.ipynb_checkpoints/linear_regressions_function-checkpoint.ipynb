```python
import seaborn as sns
sns.set(color_codes=True)
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import random

%matplotlib inline


a = [(-1, 0), (1, 3), (2, 4), (3, 7), (4, 10)]


def sloper(input):
    x = []
    y = []
    x_square = []
    xy = []
    ran = random.sample(xrange(1000), 50)# from a list of 100 randomly generated numbers, take a list of 50
    for i, j in input:
        x.append(i)
        x_square.append(i**2)
        y.append(j)
        xy.append(i*j)
    slope = (np.mean(xy)-(np.mean(x)*np.mean(y)))/(np.mean(x_square) - np.mean(x)**2)
    intercept = slope * x[1] - y[1]
    ran_y = [(i * slope)+ intercept for i in ran]
    plot_points = map(None, ran, ran_y)
    values = pd.DataFrame(plot_points)
    values.columns = ["rand_x", "rand_y"]
    plot_points2 = map(None, x, y)
    #print(type(ran), type(ran_y), type(plot_points)) #-- checking that all are equal length
    #print(values)
    #sns.regplot(x = ran, y = ran_y, data = plot_points) #randomly generate a bunch of points, save them, then plot them
    sns.regplot(x = "rand_x", y = "rand_y", data = values)
    print("The slope is {} and the y intecept is {}".format(slope, intercept))
    #
    
    
sloper(a)

#error msg: TypeError: only integer arrays with one element can be converted to an index   
 
    
```

```python
b = [(1,3.0), (2,5.1), (3, 7.1), (4, 9.4), (5, 11.5), (6,13.5), (7,15.6), (8,17.7), (9, 19.8)]

sloper(b)
```
