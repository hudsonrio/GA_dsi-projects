```python

col_list = ['weekday', 'hour', 'taxi-cs_avg_trips', 'uber_avg_trips', 'lyft_avg_trips']
trips = trips.iloc[1:, :]
trips.columns = col_list
trips = trips.applymap(lambda x: str(x))
trips = trips.applymap(lambda x: x.replace(",", ""))
# trips.isnull().sum()
```

```python


def trip_clean(string):
    try:
        return int(string)
    except:
        return 0


trips = trips.applymap(lambda x: trip_clean(x))



trips['total_trips'] = trips['taxi-cs_avg_trips'] + trips['uber_avg_trips'] + trips['lyft_avg_trips']
trips['uber_prob'] = trips['uber_avg_trips']/trips['total_trips']
trips['uber_prob']
trips = trips.iloc[:168,:]
trips['uber_prob']


def weekday_convert(day):
    if day == 1: #sunday
        return 6 #sunday in pandas
    elif day == 2: #monday
        return 0 #sunday in pandas
    elif day == 3: #monday
        return 1 #sunday in pandas
    elif day == 4: #monday
        return 2 #sunday in pandas
    elif day == 5: #monday
        return 3 #sunday in pandas
    elif day == 6: #monday
        return 4 #sunday in pandas
    elif day == 7: #monday
        return 5 #sunday in pandas

trips['weekday'] = trips['weekday'].apply(lambda x: weekday_convert(x))
trips.to_csv('non_uber_trips.csv')

trips.head()
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.cross_validation import cross_val_score, StratifiedKFold


lr_basic = LogisticRegression(solver='liblinear', penalty='l2', C=0.75) 

# df['fare_scaled'] = preprocessing.StandardScaler().fit_transform(df['Fare'])
trips['uber_taxi_bin'] = trips['uber_prob'].apply(lambda x: 1 if x>.5 else 0)
X = trips.iloc[:,0:2]

y = trips['uber_taxi_bin']
lr_basic = lr_basic.fit(X, y)
stratk = StratifiedKFold(y, n_folds=10, shuffle=True, random_state=66)
s = cross_val_score(lr_basic, X, y, cv=stratk, n_jobs=-1, scoring='f1')

s.mean()

```

```python
# from sklearn import GridSearchCV
from sklearn import svm, grid_search, datasets
logreg_full_data = LogisticRegression(solver='liblinear')

C_vals = [0.0001, 0.001, 0.01, 0.1, .15, .25, .275, .33, 0.5, .66, 0.75, 1.0, 2.5, 5.0, 10.0, 100.0, 1000.0]
penalties = ['l1','l2']

gs = grid_search.GridSearchCV(lr_basic, {'penalty': penalties, 'C': C_vals}, verbose=False, cv=15)

gs.fit(X, y)
logreg_full_data = LogisticRegression(C=gs.best_params_['C'], penalty=gs.best_params_['penalty'])
cv_log_gs_full_data = logreg_full_data.fit(X, y)

s_cv_log_gs_full_data = cross_val_score(cv_log_gs_full_data, X, y, cv=stratk, n_jobs=-1, scoring='f1')


print("{} Score:\t{:0.3} Â± {:0.3}".format("Grid Search Optimized Logistic (Ridge) Regression", s_cv_log_gs_full_data.mean().round(3), s_cv_log_gs_full_data.std().round(3)))
```
